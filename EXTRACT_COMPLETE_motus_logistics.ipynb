```python
import re
import json
import requests

import pprint 
pp = pprint.PrettyPrinter(indent=4)

import importlib
imported_module = importlib.import_module("kvt_extract_ratecon_motus_logistics")
importlib.reload(imported_module)

from kvt_extract_ratecon_motus_logistics import kvt_extract
from dukeai.rpa_functions.ratecon_schema_check import check_rate_con_schema
from dukeai.rpa_functions.ratecon_data_check import check_rate_con_data

with open('data/text_data/motus_logistics_text_data_links.json', 'r') as jb:
    text_data_files_links = json.loads(jb.read())

    
for n, text_data_link in enumerate(text_data_files_links):
    print(f"[{n}]-[Starting KVT]-{text_data_link.split('/')[-1]}\n")
    doc_text = requests.get(text_data_link).json()
#   debug:  print(doc_text)
#     print("KVT Extraction")
    text = "\n\n\n".join(doc_text['text'].values())
#    print(text)
    kvt = kvt_extract(text)
    pp.pprint(kvt)
    
#     print(kvt)
    check_rate_con_schema(kvt)
    check_rate_con_data(kvt)
#     print(json.dumps(kvt, indent=4))
#     print("\n")
    break
```
